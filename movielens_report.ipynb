```python
from movielens_analysis import Ratings, Tags

ratings = Ratings("data/ratings_small.csv")
movies = Ratings.Movies(ratings, "data/movies_small.csv")
tags = Tags("data/tags_small.csv")
```

# MovieLens Data Analysis

In this project, we analyze user behavior in a movie rating system.
We study:
- movie popularity
- rating quality
- controversial movies
- user-generated tags

Our goal is to understand how people interact with movies using real data.

## How has movie rating activity changed over time?

We analyze how the number of ratings changes across years.
This helps us understand user engagement trends.

```python
%timeit movies.dist_by_year()
movies.dist_by_year()
```

## Which movies are the most popular?

We define popularity as the number of user ratings.
More ratings usually means higher visibility and engagement.

```python
%timeit movies.top_by_num_of_ratings(5)
movies.top_by_num_of_ratings(5)
```

## Which movies have the highest average ratings?

We compute the average rating for each movie.
This shows which movies are most appreciated by users.

```python
%timeit movies.top_by_ratings(10)
movies.top_by_ratings(10)
```

## Are some movies controversial?

We measure variance in ratings.
High variance means users strongly disagree about the movie.

```python
%timeit movies.top_controversial(10)
movies.top_controversial(10)
```

## What do users usually say about movies?

Tags represent user-generated descriptions.
We analyze the most frequently used tags.

```python
%timeit tags.most_popular(10)
tags.most_popular(10)
```

## What are the longest tags?

We look at tags with the most words or characters.
These often contain detailed user opinions.

```python
%timeit tags.longest(10)
tags.longest(10)
```

## What do users associate with "love"?

We filter tags containing the word "love" to understand emotional associations in movies.

```python
%timeit tags.tags_with("love")
tags.tags_with("love")
```

# Final Insights

From this analysis, we can conclude:

- Popular movies are not always the highest rated
- Users have strong disagreements about certain movies
- Tags reveal emotional and thematic patterns in films
- Movie engagement has changed over time

This shows that user behavior in movie platforms is complex and subjective.
